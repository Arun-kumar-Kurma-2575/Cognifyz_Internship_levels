## **Task 1: Table Booking and Online Delivery**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data=pd.read_csv(r'cleaned_dataset.csv')
data.head()
```

```python
data.isnull().sum()
```

```python
X=data
import warnings
warnings.filterwarnings('ignore')
for i in X:                  # X.columns
    if X[i].dtypes=='object':
        mode=X[i].mode()[0]
        X[i].fillna(mode,inplace=True)
    else:
        mean=round((X[i].mean()),3)
        X[i].fillna(mean,inplace=True)
data.isnull().sum()   #X.isnull().sum()  
```

```python
data=data.drop('Restaurant ID',axis=1)
data.head()
```

```python
data.info()
```

```python
data['Country Code']=data['Country Code'].astype('object')
data['Price range']=data['Price range'].astype('object')
data.info()
```

### **2.1.1 Determine the percentage of restaurants that offer table booking and online delivery.**

```python
c=0
for i in range(len(data)):
    if data['Has Table booking'][i]=='Yes' and data['Has Online delivery'][i]=='Yes':
        c+=1
perc=((c/len(data))*100)
print(f'Out of {len(data)} Restaurants {c} Restaurants have both Table booking, Online delivery \nThat means {perc:.2f}% of Restaurants have both Table booking, Online delivery')
```

### **2.1.2 Compare the average ratings of restaurants with table booking and those without.**

```python
sum_of_ratings_with_table_booking=0
noof_resturants_with_table_booking=0
sum_of_ratings_wo_table_booking=0
noof_resturants_wo_table_booking=0
for i in range(len(data)):
    if data['Has Table booking'][i]=='Yes':
        sum_of_ratings_with_table_booking+=data['Aggregate rating'][i]
        noof_resturants_with_table_booking+=1
        
    else:
        sum_of_ratings_wo_table_booking+=data['Aggregate rating'][i]
        noof_resturants_wo_table_booking+=1


print(sum_of_ratings_with_table_booking,noof_resturants_with_table_booking,sum_of_ratings_wo_table_booking,noof_resturants_wo_table_booking)
Avg_rating_with_booking=((sum_of_ratings_with_table_booking)/(noof_resturants_with_table_booking))
Avg_rating_wo_booking=((sum_of_ratings_wo_table_booking)/(noof_resturants_wo_table_booking))     
print(f'Average ratings of restaurants with table booking {Avg_rating_with_booking:.3f} \nAverage ratings of restaurants without table booking {Avg_rating_wo_booking:.3f}')
```

## **Task 2: Price Range Analysis**
### **2.2.1 Determine the most common price range among all the restaurants.**

```python
d={}
for i in price_ranges:
    d[i]=0
print(d)
for i in range(len(data)):
    d[data['Price range'][i]]+=1

print(d)
print(f'Most common price range among all the restaurants is {list(d.keys())[np.argmax(list(d.values()))]}-Price Range with {np.max(list(d.values()))} no.of Restaurants')
```

### **2.2.2 Calculate the average rating for each price range.**

```python
d={}
c={}
avg_rating={}
for i in price_ranges:
    d[i]=0
    c[i]=0
    avg_rating[i]=0
print(d,c,avg_rating)
for i in range(len(data)):
    d[data['Price range'][i]]+=data['Aggregate rating'][i]
    c[data['Price range'][i]]+=1
print(d,c)
for i in price_ranges:
    avg_rating[i]=(d[i]/c[i])
    print(f'Average rating for {i}-price range={avg_rating[i]:.2f}')

```

- **Average rating for 1st price range=2.00**
- **Average rating for 2nd price range=2.94**
- **Average rating for 3rd price range=3.68**
- **Average rating for 4th price range=3.82**

### **2.3.2 Create new features like "Has Table Booking" or "Has Online Delivery" by encoding categorical variables.**
### **Encoding Some categorical columns**

```python
l=['City','Country Code','Has Table booking','Currency','Has Online delivery','Is delivering now','Switch to order menu','Rating color','Rating text']
for i in l:
    d={}
    unique=data[i].unique()
    k=0
    if ('Yes' in unique) or ('No' in unique):
        d['Yes']=1
        d['No']=0
    else:
        for j in unique:
            d[j]=k
            k+=1
    data[i]=data[i].map(d)
    print(d)
```
