### **Removing encoded text in the string to avoid confusions,randomness and saving in new dataset**

```python
import pandas as pd
import unicodedata

# Load the dataset (UTF-8 should work fine)
df = pd.read_csv(r"Dataset .csv", encoding="utf-8-sig")

# Function to normalize and clean text
def clean_text(text):
    if isinstance(text, str):
        # Normalize Unicode text (fixes special characters)
        text = unicodedata.normalize("NFKC", text)
        # Remove unknown characters (like "�")
        text = text.replace("�", "")
        # Trim spaces
        text = text.strip()
    return text

# Apply cleaning to all text columns
df = df.map(clean_text)
#df.tail()
# Save the cleaned dataset without encoding changes
df.to_csv("cleaned_dataset.csv", index=False, encoding="utf-8")
```

**Filling the Null values**
- filling missing values with **fillna()**
- Categorical columns (dtypes='object') is replaced with **mode** values of thier respective columns
- Numerical columns (dtypes!='object') is replaced with **median** values of thier respective columns

```python
X=data
import warnings
warnings.filterwarnings('ignore')
for i in X:                  # X.columns
    if X[i].dtypes=='object':
        mode=X[i].mode()[0]
        X[i].fillna(mode,inplace=True)
    else:
        mean=round((X[i].mean()),3)
        X[i].fillna(mean,inplace=True)
data.isnull().sum()   #X.isnull().sum()   
```
